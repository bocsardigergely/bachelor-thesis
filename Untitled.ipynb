```python
data = pd.read_csv('./sample.csv')[['processed_words']]
```

```python
def clean(string):
    string = string.replace('  ', '')
    string = string.replace(':', '')
    return string.split('\t')
```

```python
data['proc_words'] = data.processed_words.apply(lambda x: [clean(s) for s in x.split('\n  ')])
```
